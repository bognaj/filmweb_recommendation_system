```python
import matplotlib.pyplot as plt
import json
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
with open("return.json", "r") as f:
    data = json.load(f)
```

```python
little = []
CB_weight = []
UCS_weight = []
CF_weight = []
res = []

for ret in data:
    little.append(ret["little"])
    CB_weight.append(ret["CB_weight"])
    UCS_weight.append(ret["UCS_weight"])
    CF_weight.append(ret["CF_weight"])
    res.append(ret["res"])

little = np.array(little)
CB_weight = np.array(CB_weight)
UCS_weight = np.array(UCS_weight)
CF_weight = np.array(CF_weight)
res = np.array(res)

```

```python
fig, axs = plt.subplots(1,4)

sns.violinplot(little, ax=axs[0]);
sns.violinplot(CB_weight, ax=axs[1]);
sns.violinplot(UCS_weight, ax=axs[2]);
sns.violinplot(CF_weight, ax=axs[3]);
```

```python
fig, axs = plt.subplots(nrows=3, figsize = (10,12))

axs[0].plot(res)
axs[0].set_title("loss")
axs[1].plot(CB_weight)
axs[1].plot(UCS_weight)
axs[1].plot(CF_weight)
axs[1].set_title("weights")
axs[2].plot(little)
axs[2].set_title("little")
```

```python
data = pd.DataFrame({
    "CB_weight": CB_weight,
    "UCS_weight": UCS_weight,
    "res":res,
})

sns.scatterplot(data, x = "CB_weight", y = "UCS_weight", hue = "res")
```
